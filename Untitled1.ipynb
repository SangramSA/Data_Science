```python
from sklearn.impute import KNNImputer

# Identifying numerical columns in the dataset
numerical_columns = train_data_copy.select_dtypes(include=['int64', 'float64']).columns

# Creating the KNN imputer instance
knn_imputer = KNNImputer(n_neighbors=5)

# Applying KNN imputer to the numerical columns of the copied datasets
train_data_copy_numerical_imputed = train_data_copy.copy()
train_data_copy_numerical_imputed[numerical_columns] = knn_imputer.fit_transform(train_data_copy[numerical_columns])

test_data_copy_numerical_imputed = test_data_copy.copy()
test_data_copy_numerical_imputed[numerical_columns] = knn_imputer.transform(test_data_copy[numerical_columns])

# Displaying the first few rows of the imputed datasets
imputed_train_head = train_data_copy_numerical_imputed.head()
imputed_test_head = test_data_copy_numerical_imputed.head()

(imputed_train_head, imputed_test_head)
```

```python
# Comparing statistical properties before and after imputation for the training data
original_numerical_stats = train_data_copy[numerical_columns].describe()
imputed_numerical_stats = train_data_copy_numerical_imputed[numerical_columns].describe()

(original_numerical_stats, imputed_numerical_stats)
```

The mean, standard deviation, and the range (min to max values) of columns like Age, Work_Experience, and Family_Size are almost identical before and after the imputation. This indicates that the imputation process has been successful in maintaining the integrity of the data while filling in the missing values. ​​








```python
# Comparing statistical properties before and after imputation for the test data
original_test_numerical_stats = test_data_copy[numerical_columns].describe()
imputed_test_numerical_stats = test_data_copy_numerical_imputed[numerical_columns].describe()

(original_test_numerical_stats, imputed_test_numerical_stats)
```

This consistency indicates that the KNN imputation has effectively filled in the missing values in the test data without significantly altering the overall distribution of the numerical data. Therefore, the imputation process can be considered successful for the test data as well.

```python
from sklearn.preprocessing import OrdinalEncoder

# Identifying categorical columns in the dataset
categorical_columns = train_data_copy_numerical_imputed.select_dtypes(include=['object']).columns

# Creating an ordinal encoder instance (KNNImputer works with numerical data, so we need to convert categorical data to numerical)
ordinal_encoder = OrdinalEncoder()

# Encoding the categorical columns in the copied datasets
train_data_copy_categorical_encoded = train_data_copy_numerical_imputed.copy()
train_data_copy_categorical_encoded[categorical_columns] = ordinal_encoder.fit_transform(train_data_copy[categorical_columns])

test_data_copy_categorical_encoded = test_data_copy_numerical_imputed.copy()
test_data_copy_categorical_encoded[categorical_columns] = ordinal_encoder.transform(test_data_copy[categorical_columns])

# Applying KNN imputer to the encoded categorical columns
knn_imputer_categorical = KNNImputer(n_neighbors=5)

train_data_copy_categorical_imputed = train_data_copy_categorical_encoded.copy()
train_data_copy_categorical_imputed[categorical_columns] = knn_imputer_categorical.fit_transform(train_data_copy_categorical_encoded[categorical_columns])

test_data_copy_categorical_imputed = test_data_copy_categorical_encoded.copy()
test_data_copy_categorical_imputed[categorical_columns] = knn_imputer_categorical.transform(test_data_copy_categorical_encoded[categorical_columns])

# Decoding the imputed categorical columns back to their original categories
#train_data_copy_categorical_imputed[categorical_columns] = ordinal_encoder.inverse_transform(train_data_copy_categorical_imputed[categorical_columns])
#test_data_copy_categorical_imputed[categorical_columns] = ordinal_encoder.inverse_transform(test_data_copy_categorical_imputed[categorical_columns])

# Displaying the first few rows of the imputed datasets
imputed_train_categorical_head = train_data_copy_categorical_imputed.head()
imputed_test_categorical_head = test_data_copy_categorical_imputed.head()

(imputed_train_categorical_head, imputed_test_categorical_head)
```

```python
import matplotlib.pyplot as plt

# Function to plot the frequency distribution of categorical columns before and after imputation
def plot_categorical_distribution(column, ax):
    # Before imputation
    train_data_copy[column].value_counts(dropna=False).plot(kind='bar', ax=ax[0], title=f'{column} - Before Imputation')
    ax[0].set_xlabel(column)
    ax[0].set_ylabel('Frequency')

    # After imputation
    train_data_copy_categorical_imputed[column].value_counts(dropna=False).plot(kind='bar', ax=ax[1], title=f'{column} - After Imputation')
    ax[1].set_xlabel(column)
    ax[1].set_ylabel('Frequency')

# Creating subplots for each categorical column
fig, axes = plt.subplots(nrows=len(categorical_columns), ncols=2, figsize=(12, len(categorical_columns) * 4))

# Plotting each categorical column
for i, column in enumerate(categorical_columns):
    plot_categorical_distribution(column, axes[i])

plt.tight_layout()
plt.show()
```

From these graphs, you can observe the following for each categorical column:


*   The distributions remain fairly consistent across all the columns.

*   The imputation process has filled the missing values without significantly altering the distribution of each category.

This graphical validation indicates that the KNN imputation for categorical data has been effective and has preserved the overall structure and distribution of the categorical variables in the dataset.

#Outlier Removal

```python
# Function to remove outliers using IQR
def remove_outliers(df, column_list):
    for column in column_list:
        # Calculating IQR
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Defining the acceptable range (1.5 times IQR)
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Removing outliers
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df

# Columns to check for outliers (continuous variables)
outlier_columns = ["Gender",
        "Ever_Married",
        "Age",
        "Graduated",
        "Profession",
        "Work_Experience",
        "Spending_Score",
        "Family_Size",
        "Var_1",
        ]

# Removing outliers from the dataset
X_train_no_outliers = remove_outliers(X, outlier_columns)

# Comparing the shape before and after removing outliers
original_shape = X.shape
new_shape = X.shape

original_shape, new_shape

```
